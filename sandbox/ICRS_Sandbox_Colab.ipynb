## Step 3: View the Ranked Results

The output CSV contains: `candidate_id`, `rank`, `score`, and `reasoning`.

````python
import pandas as pd

df = pd.read_csv("sandbox_submission.csv")
print(f"Total candidates ranked: {len(df)}")
print(f"Score range: {df['score'].min():.4f} - {df['score'].max():.4f}")
print()

# Display full results with reasoning
pd.set_option('display.max_colwidth', 120)
pd.set_option('display.width', 200)
df
````

## Step 4: Score Distribution

````python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Score distribution
ax1.barh(df['rank'], df['score'], color='steelblue')
ax1.set_xlabel('Score')
ax1.set_ylabel('Rank')
ax1.set_title('Ranked Scores')
ax1.invert_yaxis()

# Score histogram
ax2.hist(df['score'], bins=15, color='steelblue', edgecolor='white')
ax2.set_xlabel('Score')
ax2.set_ylabel('Count')
ax2.set_title('Score Distribution')

plt.tight_layout()
plt.show()
````
